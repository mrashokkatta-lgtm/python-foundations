```python
# Refreshing my python3 programming skills  

# Function is a reusable block of code that performs a specific task 
```

```python
# Basic Function 

def greet():
    print("Good day Ashok!")

greet()
```

```python
# Function with parameters 

def greet(name):
    print(name)

greet(str(input("Greet a person:")))

```

```python
# Function with parameters and return  

def add(a,b):
    return a+b

res = add(3,5)
print("The sum of numbers is:",res)
```

```python
# Function with default parameters 

def add(a=3,b=2):
    return a+b 

res = add()
print("The sum of the numbers is:",res)
```

```python
# Function that returns multiple values 

def cal(a,b):
    s = a+b 
    sub = a-b
    m = a*b 
    d = a/b
    return a,sub,m,d 

q,w,e,r = cal(4,8)

print("sum of the numbers:",q)
print("Difference of the numbers",w)
print("Product of the numbers",e)
print("Division of the numbers",r)
```

```python
# Function with keyword arguments 

def add(a,b):
    return a+b 

res = add(a=2,b=3)
print("The sum of the numbers is:",res)

def add(name,age):
    print("Name of the person:",name)
    print("Age of the person:",age)

add(age=41,name="Ashok")
```

```python
# Function with variable length arguments *args 

def add_num(*args):
    total = 0
    for i in args:
        total += i 
    return total 

tot = add_num(1,2,3,4)
print("The sum of the numbers is:",tot)
```

```python
# Function with Keyword variable arguments **kwargs 

def display_info(**kwargs): 
    for key, value in kwargs.items(): 
        print(key,":", value) 

display_info(name="Ashok", age=40, interest="AI/ML")
```

```python
# Function behaving like anonymous (Lambda Function) 

square = lambda x: x*x 

print(square(10))
```

```python
# Scope of the variables : Local vs Global 

x = 5  # Global variable

def loc():
    x = 10  # Local variable
    print("The local value of x :",x)

loc()
print("The global value of x:",x)
```

```python
# practice problem : finding the factorial of a number 

def factorial(a): 
    if a <= 1:
        return 1
    else:
        return a * factorial(a-1)

print("Factorial of the number is:",factorial(5))
```
