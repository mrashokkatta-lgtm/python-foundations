# We will cover file handling in this program 


```python
# We have three types of files namely
# 1. sample.txt 
# 2. employees.csv 
# 3. config.json 
```

```python
# Opening and reading a sample.txt file 

file = open("sample.txt","r")
content = file.read()  # the entire content of the file is stored in content variable
print(content)  # printing the entire content of the file at one go
file.close()  # closing the file 
```

```python
# Reading the file content line by line 

file = open("sample.txt","r") 

for line in file:
    print(line.strip()) 

file.close()
```

```python
# best practice to open, read and close the file using "with" 

with open("sample.txt","r") as file: 
    content = file.read() 

print(content)
```

```python
with open("sample.txt","r") as file: 
    for line in file: 
        print(line.strip())
```

```python
# Writing content to a new file and then let us add more data by appending that file 
with open("newsample.txt","w") as file: 
    file.write("Daily do these 5 affirmations to achieve your dreams\n") 
    file.write("I am the best\n") 
    file.write("I can do it alone\n") 
    file.write("God is always with me\n") 

with open("newsample.txt","a") as file: 
    file.write("I am a winner\n")
    file.write("Today is my day\n")

with open("newsample.txt","r") as file: 
    line_list = file.readlines()         # printing the content of the file as list of lines. 

print(line_list)
```

```python
# Reading a .csv file 

import csv 

with open("employees.csv","r") as file: 
    content = csv.reader(file) 

    for row in content: 
        print(row)
```

```python
# writing data into .csv file 

data = {
    "name" : "Manasa", 
    "position" : "Bank Manager", 
    "salary" : 1200000
}

with open("employees.csv","a",newline="") as file: 
    fieldnames = ["name", "position", "salary"]
    writer = csv.DictWriter(file,fieldnames=fieldnames) 
    writer.writerow(data) 
```

```python
# Reading a json file 

import json 

with open("config.json","r") as file: 
    config = json.load(file) 

print(config) 
print(config["model"])
```

```python
# Writing in to a json file 

import json

data = {
    "name": "Ashok",
    "role": "ML Enthusiast",
    "skills": ["Python", "PyTorch", "Machine Learning"]
}

with open("profile.json", "w") as file:
    json.dump(data, file, indent=4)
```

```python
# Checking if files exist and get their information 

import os 

if os.path.exists("sample.txt"): 
    print("File Exists") 
else: 
    print("File not found")

print(os.getcwd())
print(os.listdir())
```
