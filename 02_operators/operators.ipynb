```python
# Refreshing my python3 programming skills 

# Arithmetic Operators

a = 10
b = 3

print("Addition:",a+b)
print("Subtraction:",a-b)
print("Multiplication:",a*b) 
print("Division:",a/b)         # returns a floating point
print("Floor Division:",a//b)  # removes decimal
print("Modulus:",a%b)          # returns a reminder
print("Exponential:",a**b) 
```

```python
# Comparison Operators

x = 5
y = 8

print("Equal:", x == y)
print("Not Equal:", x != y)
print("Greater Than:", x > y)
print("Less Than:", x < y)
print("Greater or Equal:", x >= y)
print("Less or Equal:", x <= y)
```

```python
# Logical Operators

p = False
q = False

print("AND:", p and not q)
print("OR:", not p or q)
print("NOT p:", not p)
```

```python
# Assignment Operators

# cheking for the mutability of the num variable

num = 25
print("Initial value:", num)

num += 5
print("After += :", num)

num -= 3
print("After -= :", num)

num *= 2
print("After *= :", num)

num /= 4
print("After /= :", num)
```

```python
# Membership Operators

numbers = [10, 20, 30, 40, 50]  # numbers is list data type

print("Is 30 in list?", 30 in numbers)
print("Is 80 not in list?", 80 not in numbers)
```

```python
# Identity Operators

a = [10, 20, 30]
b = a
c = [10, 20, 30]

print("a is b:", a is b)       # True (same object)
print("a is c:", a is c)       # False (different object)
print("a == c:", a == c)       # True (same values)
```

```python
# Bitwise Operators

x = 5      # 0101 binary
y = 3      # 0011 binary

print("Bitwise AND:", x & y)
print("Bitwise OR:", x | y)
print("Bitwise XOR:", x ^ y)
print("Left Shift:", x << 1)  # 0101 will be 1010 
print("Right Shift:", x >> 1) # 0101 will be 0010
```
