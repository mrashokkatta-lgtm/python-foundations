# Python Object Oriented Programming (OOP)
This notebook demonstrates basic OOP concepts:
- Class
- Object
- Constructor
- Methods
- Inheritance

```python
# Class Definition 

class Employee:

    def __init__(self,name,position,salary): 
        self.name = name 
        self.position = position 
        self.salary = salary 

    def introduce(self):
        print("Employee Name:", self.name) 
        print("Employee position:",self.position) 
        print("Employee Salary:",self.salary) 
```

```python
# Creating object to the employee class 

emp1 = Employee("Ashok", "TPM", "4000000") 
emp1.introduce()
```

```python
# Inheritance demonstration 

class Employer(Employee):

    def __init__(self,name,position,salary,company): 
        super().__init__(name,position,salary) 
        self.company = company 

    def show_company(self): 
        print(self.name,"has founded",self.company) 

# Initiating child class 
boss = Employer("Manasa","CEO",5000000,"Trading LLP")
boss.introduce()
boss.show_company()
```
