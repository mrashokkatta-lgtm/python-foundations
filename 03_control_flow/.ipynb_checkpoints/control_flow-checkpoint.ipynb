```python
# Refreshing my python3 programming skills 

# Control Flow means : Controlling the direction of program execution based on conditions 
```

```python
# Basic if statement

age = 20

if age >= 18:
    print("You are eligible to vote in India.")
```

```python
# if-else statement

age = 25

if age >= 21:
    print("You are a major")
else:
    print("You are a minor")
```

```python
# if-elif-else statement

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Grade D")
```

```python
# Nested if

age = 25
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Must be a citizen to vote")
else:
    print("Not eligible due to age")
```

```python
# Ternary operator: very useful in production code

number = 10

result = "Even" if number % 2 == 0 else "Odd"
print(result)
```

```python
# logical operators in conditions 

salary = 75000
experience = 3 

if salary > 50000 and experience >= 2:
    print("Eligible for a promotion")
else:
    print("The candidate to gain more experience for promotion")
```

```python
# Taking user input

age = int(input("Enter your age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# a sample practice test code for checking a leap year or not 

# taking year as an input 

year = int(input("Enter year to check for leap year:"))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# a sample practice test code for nested if 

# taking gender and age as inputs

gen = input("Enter gender (M/F): ").upper()
age = int(input("Enter age: "))

if age < 18:
    print("Not eligible to vote or marry.")
else:
    # Everyone 18+ can vote
    status = "eligible to vote"
    
    # Check marriage eligibility
    if gen == "F" or (gen == "M" and age >= 21):
        status += " and marriage"
    
    print(f"Eligible for: {status} as per Indian Law.")

```
