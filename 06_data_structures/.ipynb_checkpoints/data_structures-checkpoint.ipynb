```python
# Refreshing my python3 programming skills  

# Python has four built-in data structures namely list, tuple, dictionary and set each offering a unique advantage in handling data 
```

```python
# Lists are ordered, mutable collections that allow duplicates 

# Tuples are ordered, immutable collections that allow duplicates 

# Dictionary is an ordered, mutable key-value pair that does not allow duplicate keys 

# Sets are not ordered but mutable collections that does not allow duplicates 
```

```python
# Lists are initiated using square brackets 

# Tuples are initiated using parenthesis 

# sets are initiated in flower brackets 

# Dictionary is also initiated in flower brackets with colon dividing key-value pair 
```

```python
# Creating, accessing, adding and removing elements, slicing and looping ----> for lists 

num = [1,2,3,4,5,6,7,8]
vegie = ["carrot", "beetroot","cabbage","potato"]

print("The numbers in the list are:",num)
print("The list of vegetables are:",vegie) 

print("The data type of num is:",type(num))
print("The data type of vegie is:",type(vegie))

```

```python
for i in num:
    print("The numbers are:",i) 

for j in vegie: 
    print("The vegetables are:",j) 
```

```python
# append and pop are adding and deleting elements at the end of the list 

# insert and remove are adding and deleting elements as per requirements

num.append(9)
vegie.append("okra")

num.insert(0,0)
vegie.insert(0,"beans")

print(num)
print(vegie)
```

```python
num.pop()
vegie.pop()

num.remove(5)
vegie.remove("potato")

print(num) 
print(vegie)
```

```python
# Slicing the list --> indexing [start:stop:step] 

print(num[0::2])
print(vegie[1::2])
```

```python
# Tuple

# Creating, accessing ----> for Tuple 

num_t = (1,2,3,4,5,6,7,8)
vegie_t = ("carrot", "beetroot","cabbage","potato")

print("The numbers in the list are:",num_t)
print("The list of vegetables are:",vegie_t) 

print("The data type of num is:",type(num_t))
print("The data type of vegie is:",type(vegie_t))

```

```python
# demonstrating the immutableness of Tuple 

num_t[3] = 10 
vegie_t[1] = "okra" 
```

```python
# 

for index, value in enumerate(num_t):
    print(f"value at index {index} is: {value}")
```

```python
# Sets 

s1 = {1,4,3,2,5,7,6,8,9,5} 
s2 = {10,20,30,40,1,5,7,50}

print(s1) 
print(type(s1))
print(s2) 
print(type(s2))
```

```python
# set perform all the types of set operations such as 

s1.add(100)
print(s1)
print(s2)
print(s1.difference(s2))
print(s1.intersection(s2)) 
print(s1.issubset(s2))
```

```python
# set operations in another way 

print("Union of two sets:",s1 | s2)
print("Intersection of two sets:",s1 & s2)
print("Difference of two sets:",s1 - s2)
```

```python
# Dictionary : Key --> Value pair 

d = {
    "Name" : "Ashok",
    "Age"  : 40,
    "Interest" : "AI/ML",
    "Position" : "Technical Program Manager"
}

print(d)
print(d["Name"])
print(d["Position"])
```

```python
d["Interest"] = "Cyber Security"
d["salary"] = "40 LPA"
print(d)
```

```python
print(d.keys())
print(d.values())
print(d.items())
```

```python
for key,value in d.items():
    print(key,value)
```

```python
# De,onstrating nested dictionary data structure 

employees = {
    "employee1" : {"Name": "Ashok", "Position": "TPM" },
    "employee2" : {"Name": "Manasa", "Position": "Banker"}
}

print(employees["employee1"]["Name"])
print(employees)
```
